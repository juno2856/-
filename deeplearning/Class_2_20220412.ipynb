```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-2,2, step = 0.01)
y = np.log(x)
y1 = -np.log(x)
y2 = np.log(-x)
y3 = -np.log(-x)
y4 = -np.log(1-x)
```

```python
plt.figure(figsize = (6,6))
plt.plot(x,y, label = 'log(x)')
plt.plot(x,y1, label = 'y = -log(x)' )
plt.plot(x,y2, label = 'y = log(-x)')
plt.plot(x,y3, label = 'y = -log(-x)')
plt.plot(x,y4, label = 'y = -log(1-x)')
plt.axhline(0, c = 'k')
plt.axvline(0, c = 'k')
plt.axvline(1, c= 'k')
plt.ylim(-5,5)
plt.legend()

plt.show()
```
