```python
import numpy as np
import cv2
import sys
import os
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./computer_vision/fig/puppy.bmp', cv2.IMREAD_COLOR)


if img is None:
    print('img read failed')
    sys.exit
    
img_re = cv2.resize(img, (300,200), interpolation = cv2.INTER_AREA)
cv2.imwrite('image_resize.png', img_re)
    
cv2.namedWindow('img', cv2.WINDOW_AUTOSIZE)
cv2.imshow('img', img)

while True:
    Key = cv2.waitKey()
    if Key == 27 or key == ord('q'):
        break
cv2.destroyAllWindows()
```

```python
imgBGR = cv2.imread('./computer_vision/fig/puppy.bmp', cv2.IMREAD_COLOR)
imgRGB = cv2.cvtColor(imgBGR,cv2.COLOR_BGR2RGB)
imgGRAY = cv2.cvtColor(imgBGR,cv2.COLOR_BGR2GRAY)


if img is None:
    print('img read failed')
    sys.exit
    
img_re = cv2.resize(img, (300,200), interpolation = cv2.INTER_AREA)
cv2.imwrite('image_resize.png', img_re)
    

fig, axes = plt.subplots(1,3, figsize = (8,8))

axes[0].imshow(imgBGR), axes[0].axis('off')
axes[1].imshow(imgRGB), axes[1].axis('off')
axes[2].imshow(imgGRAY, cmap = 'gray'), axes[2].axis('off')
plt.show()
    
    
# cv2.namedWindow('img', cv2.WINDOW_AUTOSIZE)
# cv2.imshow('img', img)


# while True:
#     key == cv2.waitKey()
#     if key == 27 or key == ord('q'):
#         break
# cv2.destroyAllWindows()
```

```python
img_list = os.listdir('./computer_vision/fig/images1/')

img_lists = []
for i in img_list:
    file_path = './computer_vision/fig/images1/'+i
    img_lists.append(file_path)
    
print(img_lists)

cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('image', cv2.WND_PROP_FULLSCREEN,
                     cv2.WINDOW_FULLSCREEN)

idx = 0
while True:
    img = cv2.imread(img_lists[idx])
    
    cv2.imshow('image', img)
    
    if cv2.waitKey(300) == 27:
        break
        
    idx +=1
    
    if idx >= len(img_lists):
        idx = 0
        
cv2.destroyAllWindows()
```

```python
src = cv2.imread('./computer_vision/fig/fig/cow.png')
dst = cv2.imread('./computer_vision/fig/fig/green.png')
src_gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

h,w = src.shape[:2]
print(h,w)
dst_resize = cv2.resize(dst, (w,h), interpolation = cv2.INTER_AREA)

_, mask = cv2.threshold(src_gray, 240, 255, cv2.THRESH_BINARY_INV) 
# 배경이 1(흰색)이됨/INV 배경이 검은색


cv2.copyTo(src, mask, dst_resize)


if src is None or dst is None:
    print('go home')
    sys.exit()
    
cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.imshow('mask', mask)
cv2.imshow('dst_resize', dst_resize)
cv2.imshow('src_gray', src_gray)


cv2.waitKey()
cv2.destroyAllWindows()
```

```python
src = cv2.imread('./computer_vision/fig/fig/puppy.bmp')
sunglass_4 = cv2.imread('./computer_vision/fig/fig/imgbin_sunglasses_1.png',
                     cv2.IMREAD_UNCHANGED)

sunglass_4 = cv2.resize(sunglass_4, (300,150))

sunglass = sunglass_4[:, :, 0:3]
mask = sunglass_4[:,:,-1]

# print(src.shape)
# print(sunglass.shape)



if src is None or sunglass is None:
    print('image read failed')
    sys.exit()

h,w = mask.shape[:2]
crop = src[120:120+h, 220:220+w]
    
cv2.copyTo(sunglass, mask, crop)
# crop[mask >0] = (0,0,255)   #빨간색썬글라스가 나옴!
    
cv2.imshow('src', src)
cv2.imshow('sunglass', sunglass)
cv2.imshow('mask', mask)
cv2.imshow('crop', crop)

cv2.waitKey()
cv2.destroyAllWindows()
```
